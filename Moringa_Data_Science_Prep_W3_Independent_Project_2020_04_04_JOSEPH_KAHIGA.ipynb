```
#First we import the libraries we will need 

# Importing the pandas library
# 
import pandas as pd

# Importing the numpy library
#
import numpy as np
```

```
# Let's read the data from the CSV file and create the dataframe to be used
 df = pd.read_csv('Autolib_dataset (2).csv')
 df.head(5)
```

```
# Let's preview the first 5 rows of our data
# 
df.head(5)
```

```
# we first get to know more about the dataset by accessing its information
df.info()
```

```
#making the columns lower case,whitespaces, replacing 
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
df.head(5)
```

```
#looking for null values
df.isnull()
df.isnull().sum()
```

```
#describing data
df.describe()
```

```
#dropping all null values
df.drop(['displayed_comment','scheduled_at','subscription_status','status','geo_point'] ,axis=1, inplace=True)
df.head(5)
```

```
# filling missing value using fillna()   
df.fillna(0) 
```

```
x = df[(df['city'] == 'Paris') & (df['bluecar_counter'] > 0)]
df = x.groupby(['hour']).size().sort_values(ascending=False)
df
```

```
#What is the most popular hour for returning cars?
df['hour'].max()
```

```
#What station is the most popular?
df.groupby(['Address','Kind'])['Address'].count().sort_values(ascending=False).head(1)
```

```
#What station is the most popular?
#At the most popular picking hour?
df.groupby(['Public name','hour'])['Public name'].max().sort_values(ascending=False).head(1)
```

```
#What postal code is the most popular for picking up Blue cars?
df.groupby(['Cars','Postal code'])['Postal code'].max().sort_values(ascending=False)
```

```
#At the most popular picking hour?
df.groupby(['Postal code','hour'])['Postal code'].max().sort_values(ascending=False).head(1)
```

```
#Do the results change if you consider Utilib and Utilib 1.4 instead of Blue cars? 
df.groupby(['Utilib 1.4 counter','Postal code'])['Postal code'].max().sort_values(ascending=False)
```

```
#Do the results change if you consider Utilib instead of Blue cars? 
df.groupby(['Utilib counter','Postal code'])['Postal code'].max().sort_values(ascending=False)
```

```
#Identify the most popular hour of the day for picking up a shared electric car (Bluecar) in the city of Paris over the month of April 2018.
df.groupby(['month','hour'])['month'].max().sort_values(ascending=False)
```
